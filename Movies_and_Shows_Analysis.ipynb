```python
from datetime import datetime as dt
```

```python
#replacing all the missing values in the column
df["date_added"].fillna(dt(2020, 1, 1), inplace = True)
```

```python
#verifying that there are no missing values in the column
df["date_added"].isna().sum()
```

```python
#spltting the column values as per the requirement
df[["part_1", "part_2"]] = df["date_added"].str.split(", ", expand = True)
```

```python
df["genre_count"] = df["listed_in"].map(lambda x: len(x.split(",")))
```

```python
movies_by_country = df.groupby("country")["show_id"].count().reset_index().sort_values(by = "show_id", ascending = False)
```

```python
movies_by_country.loc[546]
```

```python
#concatenate is the equivalent of union in sql

second_dataset = pd.read_csv("netflix_titles_second.csv")
```

```python
second_dataset.head(5)
```

```python
new_dataset = pd.concat([df, second_dataset])
```

```python
#we can see that the number has increased from before
new_dataset.shape
```

```python
#merge method
netflix_merged = pd.merge(left = new_dataset , right = movies_by_country, how = "inner", left_on = ["country"], right_on = ["country"])
```
